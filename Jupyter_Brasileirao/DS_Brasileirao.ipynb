verificando a presença de times com mais gols feitos no ano e sua presença nos grupos (g4, g6, rebaixados)

```python
anos = brasileirao_nf_df['season'].unique()
```

```python
def verica_time_gols_em_grupo(data_in, grupo, coluna, status, anos):
    
    qtd_anos_max = len(anos)
    qtd_times_ano = 0

    for ano in anos: 
        df_ano_atual = data_in.loc[data_in['season'] == ano]
        
        if coluna == "goals_for":
            if status == "mais":
                id_maximo = df_ano_atual.goals_for.idxmax(axis=0)
                time_gols = df_ano_atual.loc[id_maximo]
            else:
                id_minimo = df_ano_atual.goals_for.idxmin(axis=0)
                time_gols = df_ano_atual.loc[id_minimo]
            
        elif coluna == "goals_against": 
            if status == "mais":
                id_maximo = df_ano_atual.goals_against.idxmax(axis=0)
                time_gols = df_ano_atual.loc[id_maximo]
            else: 
                id_minimo = df_ano_atual.goals_against.idxmin(axis=0)
                time_gols = df_ano_atual.loc[id_minimo]
        else: 
            print("coluna = goals_for ou goals_against")
        
        if grupo == "g1": 
            if time_gols.place == 1:
                qtd_times_ano+=1
                
        elif grupo == "g4":
            if time_gols.place <= 4:
                qtd_times_ano+=1
                
        elif grupo == "g6":
            if time_gols.place <= 6:
                qtd_times_ano+=1
                
        elif grupo == "rebaixados": 
            if time_gols.place >= 17:
                qtd_times_ano+=1
                
        elif grupo == "ultimo": 
            if time_gols.place == 20:
                qtd_times_ano+=1
        else:
            print("grupo = g1 ou g4 ou g6 ou rebaixados ou ultimo")
        
    return qtd_times_ano/qtd_anos_max
```

```python
def gastos_transferencias_e_grupos(data_in_br, data_in_tf, grupo, coluna, status, anos_transf):
    
    qtd_anos_max = len(anos_transf)
    qtd_times = 0
    
    for ano in anos_transf:
        transf_ano = data_in_tf.loc[data_in_tf['temporada'] == ano]
        
        br_ano = data_in_br.loc[data_in_br['season'] == ano]
        
        if coluna == "despesas": 
            if status == "mais":
                id_maximo_transf = transf_ano.despesas.idxmax(axis=0)
                elemento_time_transf = transf_ano.loc[id_maximo_transf]
            else:
                id_minimo_transf = transf_ano.despesas.idxmin(axis=0)
                elemento_time_transf = transf_ano.loc[id_minimo_transf]
                
        elif coluna == "valor_reforcos":
            if status == "mais":
                id_maximo_transf = transf_ano.valor_reforcos.idxmax(axis=0)
                elemento_time_transf = transf_ano.loc[id_maximo_transf]
            else:
                id_minimo_transf = transf_ano.valor_reforcos.idxmin(axis=0)
                elemento_time_transf = transf_ano.loc[id_minimo_transf]
        else: 
            print("status = despesas ou valor_reforcos")
        
        
        time_transf = elemento_time_transf.time
        
        time_tranf_br = br_ano.loc[br_ano['team'] == time_transf]
        
        if grupo == "g1": 
            if time_tranf_br.place.item() == 1:
                qtd_times+=1
                
        elif grupo == "g4":
            if time_tranf_br.place.item() <= 4:
                qtd_times+=1
                
        elif grupo == "g6":
            if time_tranf_br.place.item() <= 6:
                qtd_times+=1
                
        elif grupo == "rebaixados": 
            if time_tranf_br.place.item() >= 17:
                qtd_times +=1
                
        elif grupo == "ultimo": 
            if time_tranf_br.place.item() == 20:
                qtd_times+=1
        else:
            print("grupo = g1 ou g4 ou g6 ou rebaixados ou ultimo")
    
    return qtd_times/qtd_anos_max
```

```python
def pegar_posicao_time_transf(trasferencias_df, brasileiro_2015_2023, time,anos_transf):
    novo_time_atual_transf = trasferencias_df.loc[trasferencias_df['time'] == time]
    novo_time_atual_br = brasileiro_2015_2023.loc[brasileiro_2015_2023['team'] == time]
    transf_com_posicao = pd.merge(novo_time_atual_transf, novo_time_atual_br, left_on = 'temporada', right_on = 'season')
    return transf_com_posicao[['temporada','time','despesas','valor_reforcos','place']]
```

```python
def plot_grafico(transf_com_posicao):
    time_atual = transf_com_posicao['time'].unique()
    transf_com_posicao.plot(x = 'temporada',kind = 'bar', 
                     xlabel = 'Temporada',
                     ylabel = time_atual[0], grid = True)
    plt.show()
    plt.close()
```

```python
for time in times_transferencia:
    transf_com_posicao = pegar_posicao_time_transf(trasferencias_df, brasileiro_2015_2023, time, anos_transf)
    display(transf_com_posicao)
    plot_grafico(transf_com_posicao)
```

```python
campeoes_2015_2023 = brasileiro_2015_2023.query('place == 1')
display(trasferencias_df)
display(campeoes_2015_2023)
```

```python
def compara_transferencia_campeao_com_media(campeoes_2015_2023, trasferencias_df, coluna, anos_transf):
    
    qtd_anos = len(anos_transf)
    qtd_maior_media = 0
    
    for ano_campeao in  anos_transf:

        campeao_br_ano = campeoes_2015_2023.loc[campeoes_2015_2023['season'] == ano_campeao]
        campeao_df_transf_ano = trasferencias_df.loc[trasferencias_df['temporada'] == ano_campeao]
        
        media_transf_ano = campeao_df_transf_ano[coluna].mean()
        
        campeao_transf = pd.merge(campeao_br_ano, campeao_df_transf_ano, left_on = 'team', right_on = 'time')
        
        if campeao_transf[coluna].item() > media_transf_ano:
            qtd_maior_media += 1
    
    return qtd_maior_media/qtd_anos
```

```python
print(compara_transferencia_campeao_com_media(campeoes_2015_2023,trasferencias_df, 'despesas', anos_transf))
print(compara_transferencia_campeao_com_media(campeoes_2015_2023,trasferencias_df, 'valor_reforcos', anos_transf))
```
